```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data

city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
combined_data = pd.merge(city_data, ride_data)


# Display the data table for preview
combined_data.head()
```

```python
# pd.merge(ride_data,city_data,how="left")
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

urban_data = combined_data[combined_data["type"]=="Urban"]
urban_average_fares = urban_data.groupby(['city'])["fare"].mean()
urban_rides_count = urban_data.groupby(['city'])["ride_id"].count()
urban_driver_count = urban_data.groupby(['city'])["driver_count"].mean()

suburban_data = combined_data[combined_data["type"]=="Suburban"]
suburban_average_fares = suburban_data.groupby(['city'])["fare"].mean()
suburban_rides_count = suburban_data.groupby(['city'])["ride_id"].count()
suburban_driver_count = suburban_data.groupby(['city'])["driver_count"].mean()

rural_data = combined_data[combined_data["type"]=="Rural"]
rural_average_fares = rural_data.groupby(['city'])["fare"].mean()
rural_rides_count = rural_data.groupby(['city'])["ride_id"].count()
rural_driver_count = rural_data.groupby(['city'])["driver_count"].mean()

# # Build the scatter plots for each city types
plt.scatter(urban_rides_count, urban_average_fares, marker = "o", alpha = 0.5, s = 10*urban_driver_count, c = "red", label = "Urban")
plt.scatter(suburban_rides_count, suburban_average_fares, marker = "o", alpha = 0.5, s = 10*suburban_driver_count, c = "green", label = "Suburban")
plt.scatter(rural_rides_count, rural_average_fares, marker = "o", alpha = 0.5, s = 10*rural_driver_count, c = "yellow", label = "Rural")

# # Incorporate the other graph properties
plt.title(" Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (per City)")
plt.ylabel("Average Fares ($)")

# # Create a legend
plt.legend(loc = "upper right")
# # Incorporate a text label regarding circle size
plt.text(45, 30, "NOTE: Circle size correlates with driver count per city")

# # Save Figure
plt.show()
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
urban_sum_fares = urban_data["fare"].sum()
suburban_sum_fares = suburban_data["fare"].sum()
rural_sum_fares = rural_data["fare"].sum()
total_fares = urban_sum_fares + suburban_sum_fares + rural_sum_fares

urban_percent = urban_sum_fares/total_fares
suburban_percent = suburban_sum_fares/total_fares
rural_percent = rural_sum_fares/total_fares

# Build Pie Chart
labels = ["Urban","Suburban","Rural"]
sizes = [urban_percent, suburban_percent, rural_percent]
colors = ["red", "blue", "yellow"]
explode = (0.1,0,0)
plt.pie(sizes, explode = explode, labels = labels, colors = colors, autopct = "%1.1f%%", shadow = True, startangle = 270)
plt.show()

# Save Figure
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
urban_sum_rides = urban_data["ride_id"].count()
suburban_sum_rides = suburban_data["ride_id"].count()
rural_sum_rides = rural_data["ride_id"].count()
total_rides = urban_sum_rides + suburban_sum_rides + rural_sum_rides

urban_percent_rides = urban_sum_rides/total_rides
suburban_percent_rides = suburban_sum_rides/total_rides
rural_percent_rides = rural_sum_rides/total_rides


# Build Pie Chart

labels = ["Urban","Suburban","Rural"]
sizes = [urban_percent_rides, suburban_percent_rides, rural_percent_rides]
colors = ["red", "blue", "yellow"]
explode = (0.1,0,0)
plt.pie(sizes, explode = explode, labels = labels, colors = colors, autopct = "%1.1f%%", shadow = True, startangle = 270)
plt.show()

# Save Figure
```

```python
# Show Figure
plt.show()
```
